## Exercise 5.1

Write a function that reads the first 10 lines of the file `data/Homo_sapiens.GRCh38.99.MT.gtf` and writes them to another file.

````python
with open("data/Homo_sapiens.GRCh38.99.MT.gtf", mode = 'r') as infile, open("data/Homo_sapiens.GRCh38.99.MT.head.gtf", mode = 'w') as outfile:
    for _ in range(10):
        print(infile.readline().strip(), file = outfile)
````

<br>

### Solution:
Uncomment and run the cell below to show the solution.

````python
# %load solutions/solution_51.py
# Exercise 5.1

# Notes:
# To read a file with python in a Jupyter Notebook, we must specify its
# location relatively to our notebook. Since the file is located in a the
# "data" subdirectory, must add "data/" to the input file name.

input_file = "data/Homo_sapiens.GRCh38.99.MT.gtf"
output_file = "Homo_sapiens.GRCh38.99.MT.head.gtf"

with open(input_file, "r") as in_file, open(output_file, "w") as out_file:
    for i in range(10):
        print(in_file.readline(), file=out_file, end="")
        # Note: we add the optional end="" argument to the "print()" function
        # to avoid printing an additional "\n" character (newline) printed,
        # since that would cause line skips (because each line already ends
        # with a `\n` character read when read from the file).
        # Alternatively we could have used `strip()` on the lines read from
        # the input file: print(in_file.readline().strip(), file=out_file)

# Verify that the file we created above is correct by reading it.
# It should contain 10 lines.
file_content = []
with open(output_file, mode="r") as f:
    for line in f:
        file_content.append(line)

print("The file has", len(file_content), "lines")
for line in file_content:
    print(line)


# Note: in practice, reading a file line-by-line just to store its content
# in a list does not make much sense. It's simpler to use readlines() in that
# case, since we are anyway loading the whole file in memory.
with open(output_file, mode="r") as f:
    file_content = f.readlines()

print("The file has", len(file_content), "lines")
for line in file_content:
    print(line)
````


<br>

As you can see, each line is composed of 3 fields, separated by a `;`.
* The **first field** corresponds to the **strain ID**.
* The **second field** corresponds to the **geographic region** of origin of the strain.
* The **third field** corresponds to the **country**.

Furthermore, the first line does not contain data, but a **header** giving column names.

<br>

Your tasks are the following:
* Read the file and store the strain id, geographic regions, and country in three different
  lists: `strain`,`region`,`country`.
* After the file is read, print length of each list (`strain`,`region`,`country`) as well as the first
  3 elements of each list.



````python
with open("data/genbank.sars-cov2.metadata.csv", mode="r") as corona:

    corona.readline()
    strain = []
    region = []
    country = []
    for line in corona:
        ls = line.strip().split(";")
        strain.append(ls[0])
        region.append(ls[1])
        country.append(ls[2])
        # print(ls)
    print("Strain include", len(strain), "lines, e.g.", strain[:3])
    print("Strain include", len(region), "lines, e.g.", region[:3])
    print("Strain include", len(country), "lines, e.g.", country[:3])
    
````
