## Exercise 4.1 - reverse complement

**Write a function** that returns the <a href="https://en.wikipedia.org/wiki/Complementarity_(molecular_biology)">reverse complement</a> of a string containing a DNA sequence.

The reverse complement is the inverted sequence (i.e. end to start), where each nucleotide is replaced by its complement.  
Complementary nucleotides are `A` and `T`, and `G` and `C`.

Here are some examples:
* `AATGCC` -> `GGCATT`
* `AAAGGC` -> `GCCTTT`
* `AAATTT` -> `AAATTT`

Test your function with the sequence `ATAGAGCGATCGATCCCTAGCTA`, compare your results with what you can get using <a href="http://arep.med.harvard.edu/labgc/adnan/projects/Utilities/revcomp.html">this online tool</a>.

````python
seq = "AATGCC"
target = "GGCATT"

rev_seq = seq[::-1]
rev_seq
````

````python
rev_comp_seq = str()
for nuc in rev_seq:
    
    if nuc == 'T':
        print(nuc, "->", "A")
        rev_comp_seq += 'A' # rev_comp_seq.append("A")
    elif nuc == 'A':
        print(nuc, "->", "T")
        rev_comp_seq += 'T'
    elif nuc == 'C':
        print(nuc, "->", "G")
        rev_comp_seq += 'T'
    elif nuc == 'G':
        print(nuc, "->", "C")
        rev_comp_seq += 'T'    

    
    else:
        print(nuc, "// unchanged")

rev_comp_seq
````

````python
def rev_comp( seq ):
    
    rev_seq = seq[::-1]
    
    rev_comp_seq = str()
    for nuc in rev_seq:
        # use .upper() to ensure all seq contain capital letters only
        if nuc.upper() == 'T':
            #print(nuc, "->", "A")
            rev_comp_seq += 'A' # rev_comp_seq.append("A") !!! += wouldn't work with a list as such, ned to be of the same type, e.g. rev_comp_seq += ['A']
        elif nuc.upper() == 'A':
            #print(nuc, "->", "T")
            rev_comp_seq += 'T'
        elif nuc.upper() == 'C':
            #print(nuc, "->", "G")
            rev_comp_seq += 'T'
        elif nuc.upper() == 'G':
            #print(nuc, "->", "C")
            rev_comp_seq += 'T'    
        else:
            print("warning: nucleotide", nuc, "unknown >> will be ignored")
    
    return(rev_comp_seq)
````

````python
rev_comp("AATGCC")
````

````python
def rev_comp_list( *args ):
    rev_comp_list = []
    for x in args:
        rev_comp_list.append(rev_comp(x))

    return(rev_comp_list)
````

````python
rev_comp_list("AATGCC","AAAGGC","AAATTT")
````
